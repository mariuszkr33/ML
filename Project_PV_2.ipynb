```python
df_train = pd.read_csv(r'/content/drive/MyDrive/Participants_Data_WSMLC/train.csv')
df_test = pd.read_csv(r'/content/drive/MyDrive/Participants_Data_WSMLC/test.csv')
```

```python
df = pd.concat([df_train, df_test])
df.shape
```
